```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 100)
x = np.sin(t)
y = np.cos(t)
```

```python
plt.plot(t, x)
```

```python
plt.plot(t, x, 'o')
```

```python
plt.plot(t, x, 'o-')
plt.plot(t, y, 'x-')
```

```python
plt.plot(t, x, 'ro--')
plt.plot(t, y, 'x-')
```

```python
plt.plot(t, x, 'ko:')
plt.plot(t, y, 'h-', markersize = 10)
```

```python
plt.figure(figsize=(15, 7))
plt.plot(t, x, 'ko:')
plt.plot(t, y, 'h', markersize = 5)
```

```python
plt.plot(t, x, marker = '*', color = '#9090ff', markersize=20, markeredgecolor = "blue", markerfacecolor = 'yellow')
```
