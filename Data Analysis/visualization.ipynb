```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 100)
x = np.sin(t)
y = np.cos(t)
```

```python
plt.plot(t, x)
```

```python
plt.plot(t, x, 'o')
```

```python
plt.plot(t, x, 'o-')
plt.plot(t, y, 'x-')
```

```python
plt.plot(t, x, 'ro--')
plt.plot(t, y, 'x-')
```

```python
plt.plot(t, x, 'ko:')
plt.plot(t, y, 'h-', markersize = 10)
```

```python
plt.figure(figsize=(15, 7))
plt.plot(t, x, 'ko:')
plt.plot(t, y, 'h', markersize = 5)
```

```python
plt.plot(t, x, marker = 'o', markersize=20, markeredgecolor = "blue", markerfacecolor = 'yellow')
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]
prices = [120, 90, 140, 150, 180, 130]
fig = plt.figure(figsize=(5, 5))
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0, 0, .2, .2])
ax3 = fig.add_axes([0.8, 0.8, .2, .2])      # x, y width, height
ax4 = fig.add_axes([0.8, 0, .2, .2])
ax5 = fig.add_axes([0, 0.8, .2, .2])
ax6 = fig.add_axes([.3334, .4, .3, .3])
```

```python
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_axes([0, 0, 1, 1])
ax1.bar(years, prices)
```

```python
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_axes([0, 0, 1, 1])
ax1.pie(prices, labels=years, autopct='%.2f%%')
plt.show()
```
