```python
from dputils.scrape import Scraper, Tag
import pandas as pd
```

```python
url = "https://www.zyte.com/blog/"
target = Tag(cls="ResourcePage_grid__avGjs")
items = Tag(cls='CardResource_card__BhCok')
# items = Tag(cls='card')
article = Tag(cls="z-text-blog-title")

metainfo = Tag(cls='CardResource_info__g1t2c')

link = Tag('a', output='href', cls='Button_btn__WcYUU')

scr = Scraper(url)
results = scr.get_repeating_page_data(target, items, article=article, metainfo=metainfo, link=link)
df = pd.DataFrame(results)
df
```

### multipage - crawler

```python
page = 1
q = 'ssd'
all_data = []
for i in range(50):
    url = f'https://www.flipkart.com/search?q={q}&page={page}'
    print(f'getting page {page} from {url}')
    target = Tag(cls='_1YokD2 _2GoDe3')
    items = Tag(cls='_4ddWXP')
    name = Tag('a', cls='s1Q9rs')
    price = Tag(cls='_30jeq3')
    oprice = Tag(cls='_3I9_wc')
    rating = Tag(cls='_3LWZlK')
    scr = Scraper(url)
    results = scr.get_repeating_page_data(target, items, name=name, price=price, oprice=oprice, rating=rating)

    if len(results) == 0:
        print('no more results')
        break
    all_data.extend(results)
    page += 1

df = pd.DataFrame(all_data)
df
```

### database using sqlalchemy

```python
from sqlalchemy import create_engine        # database connection
import datetime as dt
from sqlalchemy.orm import sessionmaker     # session management open and close connection          # orm - object relational mapper
from sqlalchemy.orm import declarative_base         # base class for declarative class definitions
from sqlalchemy import Column, Integer, String, DateTime

Base = declarative_base()
class Product(Base):
    __tablename__ = 'products'
    id = Column(Integer, primary_key=True)
    name = Column(String)
    price = Column(String)
    oprice = Column(String)
    rating = Column(String)
    created_on = Column(DateTime, default=dt.datetime.utcnow)

    def __str__(self):
        return f'{self.name}'
    
db = 'crawler.db'
engine = create_engine(f'sqlite:///{db}')       # create database
Base.metadata.create_all(engine)        # create tables in database
```

```python
df.to_sql('products', engine, if_exists='replace')          # write to database
print('done')
```
