```python
df[['OdName', 'Total']].plot(
    x = 'OdName', y = 'Total', color = 'red',
    title='plot', figsize=(15, 5))
```

```python
df[['OdName', 'Total']].plot(
    kind = 'bar',
    x = 'OdName', y = 'Total', color = 'red',
    title='plot', figsize=(15, 5))
```

```python
df.sort_values(by='Total', ascending=False, inplace=True)
df[['OdName', 'Total']].head(25).plot(
    kind = 'bar',
    x = 'OdName', y = 'Total', color = 'green',
    title='plot', figsize=(15, 5))
```
